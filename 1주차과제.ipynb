##언어분야

* __인공지능 스포츠 기사를 작성하는 *'야알봇'*__

![야알봇 페이스북 캡쳐](https://user-images.githubusercontent.com/44537703/82758670-deaa4980-9e22-11ea-82fd-438a21b74c94.png)

야알봇은 KBO 경기 내용을 토대로 스포츠 기사를 작성하는 인공지능이다.

야알봇이 기사를 작성하는데는 4단계가 필요하다.

1. 데이터 수집

    포털 사이트에서 경기를 실시간으로 줄계하는 문자 중계 데이터를 모은다.

2. 중요 이벤트 선별

    '세이버 매트릭스' 알고리즘(다양한 수치를 활용해 중요한 플레이를 찾아내는 통계기법)을 이용하여 데이터를 추출하고, 이를 이벤트라는 이름으로 저장한다.

    이벤트에는 중요도를 부여한다.

3. 무드 탐지 

    중요 이벤트를 묶어 기사의 분위기를 설정

4. 기사 작성

    알고리즘을 이용해 비워둔 문장과 데이터에 주어와 목적어를 결합한다.

야알봇이 1500자 이내의 기사를 쓸때 걸리는 시간은 약 0.3초 라고한다. 또한 사람이 쓴 기사와 비교해봤을때, 46.9% 의 사람들은 어느것이 야알봇이 쓴 기사인지 모른다고 한다. 거의 비슷하다는 이야기다. 

단순히 빠른 시간안에 경기의 정보를 전달할때는 사람보다 야알봇을 사용하는 것이 더 좋을것 같다. 물론 오보나 예측, 감정이 실린 기사는 사람이 더 잘쓴다. 하지만 더욱더 발전된 알고리즘이 등장한다면 완벽한 스포츠 기사 뿐만 아니라 신문 기사도 인공지능이 작성하는 날이 올것이다. 매우 흥미롭다.

* __*챗봇*__

![saltlux콜봇 캡쳐](https://user-images.githubusercontent.com/44537703/82759397-2aabbd00-9e28-11ea-802d-0143cbe27b40.png)

챗봇은 음성이나 문자를 통한 인간과의 대화를 통해서 특정한 작업을 수행하도록 제작된 컴퓨터 프로그램이다. 크게 3가지 방식이 있다.

1. 자연언어처리 방식

    자연어처리 기술을 적용하는 언어이해 방식

2. 검출 방식 

    입력받은 말에서 특정 단어나 어구를 검출하여 그에 맞는 미리 준비된 응답을 출력하는 검색 방식

3. 각본 방식 

    각본을 미리 만들고, 각본에 따라서 사용자의 입력에 대한 동작과 각본에 있는 응답을 출력하는 방식

첫번째 방식인 자연어처리 방식에 대해 알아보자.

####__자연어처리__

자연어처리란 인간의 언어 형상을 컴퓨터와 같은 기계를 이용해서 모사 할수 있도롣 연구하고 이를 구현하는 인공지능의 주요 분야이다.

간단히 과정을 살펴보자.

1. 텍스트 전처리 과정

    문자을 용도에 맞게 가공하는 작업이다. 토큰화라는 과정을 커쳐 문장을 잘게 쪼개고 의미가 있는 토큰과 그렇지 않는 토큰을 알맞게 분리한다.

2. 언어 모델 선택

    언어의 종류 별로 문법과 어순이 다르므로 알맞은 모델을 선택해야한다. 여기서 모델의 역활은 묶인 단어들이 얼마나 연관성이 있는가를 확률로 할당하는 역활을 한다.

    즉, 단어 뒤에 어떤 단어가 올까?에 대해 확률적으로 계산을 하는 모델이다.

3. 토픽 모델

    글을 쓰거나 말을 할때는 주제(topic)에 맞게 해야한다. 토픽 모델은 단어들을 묶어 어떤 주제가 얼마나 포함됬는지를 알려준다.

4. Word Embedding

    각각 단어별의 연관성을 정해준다. 단어들을 벸터로 표현하여 처리한다.

매우 간단히 설명했다. 각 모델의 종류도 많고 표현방법도 여러가지가 있다. 앞으로 더 배워보고 싶다.

자연어처리를 이용한 챗봇은 매우 많은 곳에 쓰인다. NH 농협 고객상담 응답서비스도 그중 하나이다. 
* 사용자와 주고 받은 내용을 가지고 인공지능이 스스로 답변 제시
* 상담사와 고객의 내용을 가지고 인공지능이 적걸한 답변을 제공, 이을 상담사가 참고하여 빠른 대응.

##음성분야

인공지능의 음성 분야라 하면 크게 두가지가 있다.

* __음성 인식__(Speech Recognition)
    
    사람이 말하는 음성 언어를 컴퓨터가 해석해 그 내용을 문자 데이터로 전환하는 처리를 말한다. STT(Speech-to-Text)라고도 한다.

* __음성 합성__(Speech synthesis)
    
    말소리의 음파를 기계가 자동으로 만들어 내는 기술, TTS(Text-to-Speech)라고도 한다.

각각의 기술들을 간단히 살펴보자

####__음성인식__
![음성인식기 기본구조](https://user-images.githubusercontent.com/44537703/82761103-13be9800-9e33-11ea-8ed9-8bb41c85662c.png)

* 음형 모델

    우리말의 ㄱ, ㄴ, ㄷ, ㄹ, ..., ㅏ, ㅑ, ㅓ, ㅕ, ... 등의 소리단위를 딥러닝 기술로 학습하여 들려오는 소리를 변환한다.

* 언어 모델

    자연어처리 기술과 같다. 단어나 어휘가 어느정도 확률로 오는가를 알려주는 통계 모델

 위의 두 모델을 이용하여 들리는 말소리를 문자로 변형시킨다.

####__음성합성__

음성합성 기술은 상당히 오래되었다. 최근 10여 년 사이 딥러닝 기술을 활용해 비약적인 성능향상을 이뤄냈다.

딥러닝 기반 음성합성 기술은 3세대로 볼 수 있다.

1. 1세대(2013)

    딥러닝 기술을 음성합성 분야에 활용하려 한 세대. 딥러닝의 심층 신경망(DNN)을 이용했다.

2. 2세대(2015)

    예측에 뛰어난 성능을 보이는 LSTM-RNN(Long Short Term Memory Recurrent Neural Network)을 활용하였다. 음성 파라미터 자체를 직접 예측하려한 시기이다.

3. 3세대(2017~)

    음성의 특징 파라미터들을 대신해 음성 신호를 직접 예측하기 시작했다. 또한 1, 2 세대에서 각각의 모듈과 각 단계별로 전문 지식이 필요했는데 3세대에서는 입력부터 출력까지 하나의 모듈로 구성되었다. 

    입력 텍스트와 그에대한 음성 데이터만 가지고도 모델을 학습할 수 있게 되었다.

##이미지 분야

인공지능의 발전으로 컴퓨터는 이미지(객체)를 인식할 수 있게되었다. 이미지 인식에는 다양한 접근 방식을 사용할 수 있다. 최근에는 머신러닝과 딥러닝 기술이 이미지 인식 문제에 널리 사용되고 있다. 간단히 알아보자. 

![객체 인식에 사용되는 머신 러닝과 딥러닝 기술(자료:매스웍스)](http://www.aitimes.kr/news/photo/201807/12087_12004_574.png)

* 딥러닝

    딥러닝 모델은 객체를 식별하기 위해 해당 객체 고유의 특징을 자동으로 학습하는데 사용된다. 수많은 훈련용 이미지를 통해 특징을 학습하여 객체를 식별할 수 있다. 

* 머신러닝

    머신러닝 모델에서는 훈련용 이미지에서 주요 특징은 선택해야한다. 선택된 특징을 머신러닝 모델에 추가하여 각 특징을 고유 카테고리로 나눈다. 이 특징은 새로운 객체를 분석 및 분류할 때 사용된다. 정확한 객체 인식 모듈을 만들기 위해서는 다양한 머신러닝 알고리즘과 특징 추출 방법을 조합해서 사용해야 한다.

> 머신러닝과 딥러닝의 차이

이미지를 분류할 특징을 명확하게 알고 있다면 머신러닝을 사용하는 편이 좋다. 적은 훈련데이터로도 효과적이기 때문이다. 하지만 레이블이 지정된 대량의 데이터와 고성능 GPU가 있다면 딥러닝을 사용하는게 편하다. 딥러닝은 특징을 스스로 찾아내기 때문이다.

##자율주행

자율주행은 말 그대로 스스로 주행하는 기술이다. 자율주행하면 자연스럽게 자율주행차가 떠오리지만, 실제론 비행기, 드론, 선박 심지어는 가전제품에도 사용되고 있다. 하지만 인공지능과 자율주행 기술이 결합되면서 자율주행차가 빠르게 발전되고 있다. 자륭주행차에서는 인공지능이 어떤 역활을 하는지 알아보자.

* 보행자 충돌 경고

    차량 운행 도중에 등장하는 보행자를 감시하고 회피하는 기능

* 차선이탈 경고

    차량 운행 도중에 차선이탈을 감지하고 경고하는 기능

* 전방충돌경보

    차량 운행 도중에 앞에 나타난 차량이나 장애물을 감지하고 경보하는 기능

* 속도제한 경고

    운행 중 속도 제한 표지판을 읽어 경보하는 기능

* 교통신호 감지

    각종 신호등과 표지판을 읽어 표시하는 기능

위의 기능을 구현하기 위해 인공지능을 도입했다. 특히, 딥러닝을 도입하였다. 딥러닝을 도입한 이유는 첫째, 사람들이 데이터에 의미를 부여하지 않는 비지도학습이 가능하다. 둘째, 사람의 개입없이 특징을 추출할 수 있다. 셋째, 매 순간마다 인공신경망을 쌓아올리기 때문에 학습에 용이하기 때문이다. 

이미지, 영상 처리 뿐만이 아니라 라이더센서에서 얻은 정보를 통햐 차량 주변 환경을 더 정확하게 파악하기위해 인공지능을 필수로 사용한다.
